```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dff = pd.read_csv("PurchasesFINAL12312016.csv")
```

```python
plt.figure(figsize=(5,10))
plt.barh(dff["Size"],dff["Quantity"])
```
